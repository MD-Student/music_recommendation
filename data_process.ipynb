# 数据预处理

```python
import pandas as pd
import missingno as msno

df_train = pd.read_csv("data/train.csv")
df_members = pd.read_csv("data/members.csv")
df_songs = pd.read_csv("data/songs.csv")
```

```python
# 可视化三个表中每一列数据的缺失情况
msno.matrix(df_train, labels=True)
msno.matrix(df_members, labels=True)
msno.matrix(df_songs, labels=True)
```

```python
# 计算每一列的缺失率
missing_rate = (df_train.isnull().sum() / df_train.shape[0]).sort_values(ascending=False)
missing_rate = missing_rate.map(lambda x: "{:.2%}".format(x))
df_missing_rate = pd.DataFrame({
    'column': missing_rate.index,
    'missing_rate': missing_rate.values
})
display(df_missing_rate)
```
